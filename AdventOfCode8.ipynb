```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from geopandas import GeoDataFrame
import shapely
from shapely.geometry import Point
import plotly.express as px
```

```python
df = pd.read_csv('input8.csv', header=None)
row = df[0][0]
for i in range(len(df[0])):
    df[0][i] = [*df[0][i]]
vis = pd.DataFrame([[0]*len(df[0][0])]*len(df[0]))
```

```python
#Set vis
for a in range(len(vis)):
    for b in range(len(vis[a])):
        if a == 0 or b == 0:
            vis[a][b] = 1
        if a == len(vis)-1 or b == len(vis[a])-1:
            vis[a][b] = 1
print(vis)
```

```python
#Left to right
for a in range(len(df[0])):
    curr = -1
    for b in range(len(df[0][a])):
        mov = int(df[0][a][b])
        if mov > curr:
            curr = int(df[0][a][b])
            vis[b][a] = 1
        else:
            null = 0
            
#Right to left
for a in range(len(df[0])-1,-1,-1):
    curr = -1
    #a = len(df[0])-a-1
    for b in range(len(df[0][a])-1,-1,-1):
        mov = int(df[0][a][b])
        if mov > curr:
            curr = int(df[0][a][b])
            vis[b][a] = 1
        else:
            null = 0

#Up to down
for b in range(len(df[0][0])):
    curr = int(-1)
    for c in range(len(df[0])):
        mov = int(df[0][c][b])
        if (mov>curr):
            curr = int(df[0][c][b])
            vis[b][c] = 1
        else:
            null = 0

#Down to up
for b in range(len(df[0][a])-1,-1,-1):
    curr = -1
    for c in range(len(df[0])-1,-1,-1):
        mov = int(df[0][c][b])
        if mov > curr:
            curr = int(df[0][c][b])
            vis[b][c] = 1
        else:
            null = 0
```

```python
# Part 2
scores = pd.DataFrame([[0]*len(df[0][0])]*len(df[0]))
#Left to right
for i in range(len(df[0])):
    for j in range(len(df[0][i])):
        left = 0
        right = 0
        up = 0
        down = 0
        #print(i,j)
        curr = int(df[0][i][j])
        top = 0
        for b in range(j+1,len(df[0][i])):
            mov = int(df[0][i][b])
            if mov < curr:
                right += 1
                print(mov>=top, mov<curr, mov)
                top = mov
            else:
                right +=1
                #print("R",right)
                break
        #print("R1",right)
        
        
        #Right to left
        #curr =  int(df[0][i][j])
        #a = len(df[0])-a-1
        top = 0
        for b in range(j-1,-1,-1):
            mov = int(df[0][i][b])
            #print(i,b,mov)
            if mov <curr:
                top = mov
                left += 1
                #print("L",left)
            else:
                left += 1
                #print("Left",left)
                break
        #print("L1",left)
        
        
        #Up to down
        
        curr =  int(df[0][i][j])
        top = 0
        for c in range(i+1,len(df[0])):
            mov = int(df[0][c][j])
            if mov <curr:
                top = mov
                #curr = int(df[0][c][j])
                down +=1
            else:
                down +=1
                #print("Down", down)
                break
        #print("D1",down)
        
        
        #Down to up
        
        curr =  int(df[0][i][j])
        top = 0
        for c in range(i-1,-1,-1):
            mov = int(df[0][c][j])
            if mov <curr:
                top = mov
                #curr = int(df[0][c][j])
                up +=1
                #print("U,",up)
            else:
                up +=1
                #print("Up", up)
                break
        #print("Up1", up)
        score = left*right*down*up
        scores[j][i] = score
        print(i,j,score)
print(scores)
```
