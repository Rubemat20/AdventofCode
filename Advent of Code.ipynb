```python
f = open('input2.txt', 'r')
df = pd.DataFrame(columns=['Game', 'inputs', 'maxRed', 'maxGreen', 'maxBlue'])
for line in f:
    newRow = pd.Series({"Game":line[4:line.index(':')], "inputs":line[line.index(':')+1:], "maxRed":0,"maxGreen": 0,"maxBlue": 0})
    df = pd.concat([df,newRow.to_frame().T])
df = df.set_index("Game")
df = df.reset_index()
```

```python
for i in range(len(df.Game)):
    draws = df.inputs[i].split(';')
    maxRed = 0
    maxGreen = 0
    maxBlue = 0
    for single in draws:
        for draw in single.split(', '):
            if (draw[0] == " "):
                draw = draw[1:]
            if ("green" in draw):
                curGreen = draw[:draw.index(" ")]
                if int(curGreen) > maxGreen:
                    maxGreen = int(curGreen)
            if ("blue" in draw):
                curBlue = draw[:draw.index(" ")]
                if int(curBlue) > maxBlue:
                    maxBlue = int(curBlue)
            if ("red" in draw):
                curRed = draw[:draw.index(" ")]
                if int(curRed) > maxRed:
                    maxRed = int(curRed)
    df.maxRed[i] = maxRed
    df.maxGreen[i] = maxGreen
    df.maxBlue[i] = maxBlue
df
```

```python
# Check 12 red, 13 green, 14 blue
total = 0
for i in range(len(df.index)):
    if (df.maxRed[i] <= 12) and (df.maxGreen[i] <= 13) and (df.maxBlue[i] <= 14):
        total += i+1
#total
```

```python
f = open('input3.txt', 'r')
count = 0
matrix = np.empty((140, 140), dtype='U1')
for line in f:
    charCount = 0
    for char in line[:len(line)-1]:
        matrix[count][charCount] = char
        charCount += 1
    count += 1
#Puts the input into a matrix format
matrix
```

```python
def checkForParts(i, j):
    for iCheck in range(max(i-1, 0), min(i+2, 139)):
        for jCheck in range(max(j-1, 0), min(j+2, 139)):
            if not matrix[iCheck][jCheck].isdigit():
                if matrix[iCheck][jCheck] != ".":
                    return True
    return False
```

```python
total = 0
for i in range(140):
    j = 0
    while j < 140:
        if matrix[i][j].isdigit():
            k = 1
            val = matrix[i][j]
            part = checkForParts(i, j)
            while ((j+k) < 140) and matrix[i][j+k].isdigit():
                val = val + matrix[i][j+k]
                if part == False:
                    part = checkForParts(i, j+k)
                k += 1
            val = int(val)
            if part:
                total += val
            j += k
        else:
            j += 1
#total
```

# Day 3: Part 2

```python
def checkForGears(iCheck, jCheck):
    for a in range(max(iCheck-1, 0), min(iCheck+2, 139)):
        for b in range(max(jCheck-1, 0), min(jCheck+2, 139)):
            if matrix[a][b] == "*":
                return (True, (a, b))
    return (False, (0, 0))
```

```python
gears = []
for i in range(140):
    j = 0
    while j < 140:
        part = False
        if matrix[i][j].isdigit():
            k = 1
            val = matrix[i][j]
            part, spot = checkForGears(i, j)
            while ((j+k) < 140) and matrix[i][j+k].isdigit():
                val = val + matrix[i][j+k]
                if part == False:
                    part, spot = checkForGears(i, j+k)
                k += 1
            val = int(val)
            if part:
                valAdded = False
                for g in range(len(gears)):
                    if gears[g][0] == spot[0] and gears[g][1] == spot[1]:
                        gears[g].append(val)
                        valAdded = True
                if not valAdded:
                    gears.append([spot[0], spot[1], val])
            j += k
        else:
            j += 1
total = 0
for gear in gears:
    if len(gear) == 4:
        total += gear[2] * gear[3]
#total
```

```python
def tiltNorth(a, b):
    while a>0:
        if newMat[a-1][b] == "O":
            return
        if newMat[a-1][b] == ".":
            newMat[a-1][b] = "O"
            newMat[a][b] = "."
        if newMat[a-1][b] == "#":
            return
        a = a-1
def tiltSouth(a, b):
    while a<99:
        if newMat[a+1][b] == "O":
            return
        if newMat[a+1][b] == ".":
            newMat[a+1][b] = "O"
            newMat[a][b] = "."
        if newMat[a+1][b] == "#":
            return
        a = a+1
def tiltWest(a, b):
    while b>0:
        if newMat[a][b-1] == "O":
            return
        if newMat[a][b-1] == ".":
            newMat[a][b-1] = "O"
            newMat[a][b] = "."
        if newMat[a][b-1] == "#":
            return
        b = b-1
def tiltEast(a, b):
    while b<99:
        if newMat[a][b+1] == "O":
            return
        if newMat[a][b+1] == ".":
            newMat[a][b+1] = "O"
            newMat[a][b] = "."
        if newMat[a][b+1] == "#":
            return
        b = b+1
```

```python
high = 0
low = 1
for x in range(1):
    signals = [[file[broadcaster][1].copy(), "l", "roadcaster"]]
    while len(signals) > 0 and len(signals) <30:
        cur = signals.pop(0)
        curSig = cur[1]
        if curSig == "l":
            low += 1
        else:
            high += 1
        while len(cur[0])>0:
            new = cur[0].pop(0)
            check = 0
            if new == "rx":
                break
            while file[check][0][1] != new:
                check += 1
            if file[check][0][0] == "%" and curSig != "h":
                if file[check][2] == 0:
                    file[check][2] = 1
                    signals.append([file[check][1].copy(), "h", file[check][0][1]])
                else:
                    file[check][2] = 0
                    signals.append([file[check][1].copy(), "l", file[check][0][1]])
            if file[check][0][0] == "&":
                inputLen = file[check][3]
                count = 0
                for into in range(len(file[check][3])):
                    if file[check][3][into][0] == cur[2]:
                        file[check][3][into][1] = curSig
                    if file[check][3][into][1] == "h":
                        count += 1
                if inputLen == count:
                    signals.append([file[check][1].copy(), "l", file[check][0][1]])
                else:
                    signals.append([file[check][1].copy(), "h", file[check][0][1]])
print(high)
print(low)
```
